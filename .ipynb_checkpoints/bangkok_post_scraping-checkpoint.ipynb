```python
# click "LOAD MORE" few times to include all news

waiting_time = 15
# browser.get_window_size() 
half_monitor_size = 300 # depend on the monitor size which current using
# loop_times = 0
loop_times = 10

time.sleep(waiting_time) # initial wait 

for x in range(loop_times):
#     browser.implicitly_wait(waiting_time)
    button = WebDriverWait(browser,waiting_time).until(EC.presence_of_element_located((By.LINK_TEXT, 'LOAD MORE')))
    button_loc = button.location['y']
    script = "window.scrollTo(0, " + str(button_loc-half_monitor_size) + ");"
    browser.execute_script(script)
    button.click()
    time.sleep(waiting_time)  #wait while loading

# time.sleep(waiting_time)    
button = WebDriverWait(browser,waiting_time).until(EC.presence_of_element_located((By.LINK_TEXT, 'LOAD MORE')))
button_loc = button.location['y']    
script = "window.scrollTo(0, " + str(button_loc-half_monitor_size) + ");"
browser.execute_script(script)

# use bs4 to extract
soup = BeautifulSoup(browser.page_source, 'html.parser')
# get every entries from the main page
entries = soup.find_all(class_='listnews-row') 


# get summary
print(len(entries), " entries in the webpage\n")
timestamp = entries[-1].find_all(class_="listnews-datetime")[0].get_text()
print("date of last article:", timestamp)
print()
# test get title
print("title of the last entry:",entries[-1].find_all('h3')[0].find_all('a')[0].get_text())
# test get link
link_prefix='https://www.bangkokpost.com'
last_link_suffix=entries[-1].find_all('h3')[0].find_all('a')[0]['href']
last_link = link_prefix + last_link_suffix
print(last_link)
```

```python
# get all links and titles
links = []
titles = []
timestamps = []
link_prefix='https://www.bangkokpost.com'
waiting_time = 15

for entry in entries:
    #get title
    title = entry.find_all('h3')[0].find_all('a')[0].get_text()
    titles.append(title)
    
    #get link but exclude learning articles and video articles
    link_suffix=entry.find_all('h3')[0].find_all('a')[0]['href']
    link = link_prefix + link_suffix
    if re.search('learning',link) or re.search('vdo',link):
        link=''
    links.append(link)
    
    #get timestamp of the entry
    timestamp = entry.find_all(class_="listnews-datetime")[0].get_text().rstrip()
    if re.search(':',timestamp):
        timestamp = date.today().strftime("%d %b %Y")
    timestamps.append(timestamp)
    


print(len(timestamps))

print(len(links))
#links
# timestamps
# timestamps[-45]
```

```python
# go into every link, and get text contents of news 
texts = []

for number,link in enumerate(links):
    # time.sleep(waiting_time) 
    text = ''
    print(number+1," ", end='')
    print(link)
    # skip learning articles
    if not link =='':
        page = requests.get(link) 
        soup = BeautifulSoup(page.content, 'html.parser')
        article =soup.find_all(class_='articl-content')[0].find_all('p')
        
        for paragraph in article:
            text = text  + paragraph.get_text() +' \n\n '
        
    texts.append(text)
```

```python
# set filename
WeekNumber = date.today().isocalendar()[1]
Filename = 'W'+f"{WeekNumber:0>2}"  + '_' + 'Record.xlsx'




# create dataframe with the recent news
recent_news = pd.DataFrame({'Title':titles, 'Link':links, 'Content':texts,'Date':timestamps})


# read existing file if it exists, and append news to it
if  os.path.isfile(Filename):
    oldnews = pd.read_excel(Filename)
    print("read existing file")
    combine_news = oldnews.append(recent_news)

else:
    print("new file created from new week.")
    combine_news = recent_news

# append news to the existing file
# combine_news = oldnews.append(recent_news)

# drop duplicate and reset index
combine_news.drop_duplicates(subset='Title', keep='last', inplace=True)
combine_news.reset_index(inplace=True)
combine_news.drop(labels='index', axis=1, inplace=True)

combine_news



writer = pd.ExcelWriter(Filename)
combine_news.to_excel(writer, index=False)
writer.save()
```

```python

# ISSUE:  stopword list doesn't work for some words, e.g. "Thai" "Thailand"


# Generate a word cloud
stw = set(STOPWORDS)
stopwords = ['said','will','month','Thai','Thailand','one','two','according to','Mr','many','including','on Tuesday']
stw.update(stopwords)

wc = WordCloud(stopwords=stw ,background_color="white", max_words=40).generate(all_text)
# wc.words_

# manually apply stopwords again
for i in stopwords:
    wc.words_.pop(i, None)
# wc.words_.pop('people',None)
# dic1 = wc.words_
wc2=WordCloud(width = 800, height= 400, background_color="white", max_words=20).generate_from_frequencies(wc.words_)
print(wc2.words_.keys())


# Display the generated image:
plt.figure(figsize=(10,10))
plt.imshow(wc2, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# input

text="""Hello Mr. Smith, how are you doing today? The weather is great, and city is awesome.
The sky is pinkish-blue. You shouldn't eat cardboard"""
# text=input()
text=all_text.lower()

# tokenized
from nltk.tokenize import word_tokenize
tokenized_word=word_tokenize(text)
# print(tokenized_word)

# apply Multi-Word Expression tokenizer
from nltk.tokenize import MWETokenizer
# m_tokenizer = MWETokenizer()
m_tokenizer = MWETokenizer([('prime', 'minister')])
# m_tokenizer.add_mwe(('the', 'west', 'wing'), )
tokenized_word=m_tokenizer.tokenize(tokenized_word)



# stopword 
# modify the list by adding words of your choice in the english .txt. file in the stopwords directory.
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
filtered_text=[]
for w in tokenized_word:
    if w not in stop_words:
        filtered_text.append(w)
# print("Tokenized text:",tokenized_word)
# print("Filterd text:",filtered_text)




# word frequency
from nltk.probability import FreqDist
fdist = FreqDist(filtered_text)
# print(fdist)
fdist.most_common()

# Frequency Distribution Plot
import matplotlib.pyplot as plt
fdist.plot(30,cumulative=False)
plt.show()

# TODO: count prayut and prime_minister as same word

# word cloud
wc2=WordCloud(width = 800, height= 400, background_color="white", max_words=100).generate_from_frequencies(fdist)
# print(wc2.words_.keys())


# Display the generated image:
plt.figure(figsize=(10,10))
plt.imshow(wc2, interpolation='bilinear')
plt.axis("off")
plt.show()
```
